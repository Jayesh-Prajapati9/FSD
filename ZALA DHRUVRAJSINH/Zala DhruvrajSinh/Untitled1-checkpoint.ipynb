```python
# write pajer program your solution should prompt for a file name and display the text file 5 lines at a time pausing
# each time to ask the user to press any key to continue and q to quit 
f=open(input("Enter The File Name:- "))
for i,line in enumerate(f):
    if i%5==0 and i!=0:
        x=input("Enter any key to continue, q to quit:- ")
        if x.lower()=='q':
            break
        else:
            print(line)
    else:
        print(line)
```

```python
# write a program to find longest word or words from the given text file
f=open("pajer.txt","r")
max=0
l=list()
for i in f:
    a=i.split()
    for j in a:
        if len(j)>max:
            max=len(j)
        if j in l:
            continue
        else:
            l.append(j)
for i in l:
    if len(i)==max:
        print(i)
```

```python
#write a program to compare two text files if they are diffrent give the line n character number in the files where the
#first diffrence occurs 
f1=open("f1.txt","r")
f2=open("f2.txt","r")
file1=f1.readlines()
file2=f2.readlines()
for i,line in enumerate(file1):
    if line!=file2[i]:
        for j,char in enumerate(line):
            if char!=file2[i][j]:
                print("line:-",i+1,"character:-",j+1)
                break
        break
```

```python
#you are given a file called logfile.txt that has log on and log off times for the user each line has three entries
#seperated by commas 1 is username 2 is logon time and 3 is logoff time time is n 24hrs time format you may assume that all
#logon and logoff occur within a day.wap that scans through the file and prints out who were online
#for atleast an hour with that online time
f=open("f1.txt","r")
for i in f:
    l=i.split(",")
    x=l[1]
    y=l[2][0:5]
    time1=(int(x[0:2])*60)+(int(x[3:5]))
    time2=(int(y[0:2])*60)+(int(y[3:5]))
    if time2-time1>=60:
        print(l[0],"With The Time:-",(time2-time1)/60,"Hours")
```

```python
# write a program that prompts for a file name and look for the lines in below format:
# Extract the floating point from each of this type of lines and compute the averaje of those values and produce the output
f=open(input("Enter The File Name:- "),"r")
floating_sum=0.0
count=0
for i in f:
    if ("X-DSPAM-Confidence:") in i:
        floating_sum+=float(i[20:])
        count+=1
print("Average:-",(floating_sum/count))
f.close()
```

```python
f=open("mbox-short.txt","r")
d=dict()
for i in f:
    if i.startswith("From "):
        address=i.split()
        d[address[1]]=d.get(address[1],0)+1
max=0
for i in d:
    if d.get(i)>max:
        name=i
        max=d.get(i)
print(name,":-",max)
```

```python
f=open("mbox-short.txt","r")
d=dict()
for i in f:
    if i.startswith("From "):
        a=i.split()
        x=a[5][0:2]
        d[x]=d.get(x,0)+1
d=dict(sorted(d.items()))
for i in d:
    print(i,d.get(i))
```

```python
f = open("mbox-short.txt","r")
d = dict()
for s in f:
    for i in s.split():
        if i[0] not in d.keys():
            d[i[0]]=[i]
        else:
            d[i[0]].append(i)
f.close()
print(d)
```
